```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image

for i in range(10):
    file_prefix = f"../data/{i}"
    plt.imshow(Image.open(file_prefix + ".jpg"))
    plt.title("Image")
    plt.show()
    
    plt.imshow(np.load(file_prefix + ".npy"))
    plt.title("Real palette")
    plt.show()
    
    plt.imshow(extract_palette(file_prefix + ".jpg"))
    plt.title("Colorgram palette")
    plt.show()
```
