```python
import pandas as pd
import numpy as np
import torch

from skimage.color import rgb2lab, lab2rgb
from matplotlib import pyplot as plt

from repalette.datasets import PairRecolorDataset
from repalette.utils.normalize import Scaler

%matplotlib inline
```

```python
multiplier = 8
dataset = PairRecolorDataset(multiplier)
data_loader = torch.utils.data.DataLoader(dataset, batch_size=multiplier, num_workers=0)
```

```python
batch = next(iter(data_loader))
```

```python
scaler = Scaler()
```

```python
fig, ax = plt.subplots(nrows=8, ncols=2, figsize=(8, 40))
for i in range(4):
    ax[2*i, 0].imshow(lab2rgb(scaler.inverse_transform(batch[0][0][i]).permute(1, 2, 0)))
    ax[2*i+1, 0].imshow(lab2rgb(scaler.inverse_transform(batch[0][1][i]).permute(1, 2, 0)))
    ax[2*i, 1].imshow(lab2rgb(scaler.inverse_transform(batch[1][0][i]).permute(1, 2, 0)))
    ax[2*i+1, 1].imshow(lab2rgb(scaler.inverse_transform(batch[1][1][i]).permute(1, 2, 0)))
plt.show()
```
